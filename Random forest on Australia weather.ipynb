```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
for i in df1.columns:
  df1[i] = lbl.fit_transform(df1[i])
```

```python
from sklearn.impute import SimpleImputer
si  = SimpleImputer(missing_values = np.nan,strategy  = "mean")
df1.iloc[:,:] = si.fit_transform(df1.iloc[:,:])
```

```python
X = df1.drop(['Date','RainTomorrow'],axis =1)
Y = df1['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=0,test_size=.2)
```

```python
'''from sklearn.preprocessing import StandardScaler
strd = StandardScaler()
x_train = strd.fit_transform(x_train)'''
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=160,max_depth =10,criterion= 'log_loss',n_jobs= -1,max_features= 20)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
